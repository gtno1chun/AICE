# Chapter 1 

## 1. 머신러닝의 이해 

### 1-(2) 머신러니의 종류 

- 지도학습 : 입력 데이터와 함께 정답(라벨)이 제공되며, 이를 기반으로 모델을 학습합니다. 
- 비지도학습 : 정답이 없는 데이터를 기반으로 데이터 내 숨겨진 패턴을 학습합니다.
- 강화학습 : 에이전트가 환경과 상호작용하며 보상을 최대화하는 방법을 학습합니다.

### 1-(4) 머신러닝이 동작하는 과정 
데이터 준비 -> 모델 학습 -> 예측 -> 평가 및 개선 

## 2.딥러닝의 이해 

### 2-(2) 딥러닝의 종류 
- 합성곱 신경망(CNN, Convolutional Neural Networks) : 이미지 처리에 특화된 딥러닝 모델. (예: 얼굴인식, 자율주행 차량의 객체 인식) 
- 순환 신경망(RNN, Recurrent Neural Networks) : 시간적 순서가 중요한 데이터를 학스. (예: 인식, 텍스트 번역, 주가 예측) 
- 생섣적 적대 신경망(GAN, Genrative Adversarial Networks) : 새로운 데이터를 생성하는데 사용. (예: 가짜 이미지 생성, 예술작품 스타일 변환) 

### 2-(4) 딥러닝의 동작 과정 
데이터 준비 -> 신경망 설계(입력층, 은닉층, 출력층) -> 모델학습 -> 예측 -> 평가 및 개선 

문제 : 딥러닝이 머신러닝과 비교하여 가지는 주요 차이점은 무엇인가? 
:: 딥러닝은 인공신경망(ANN) 구조를 기반으로 여러 층(layer)으로 데이터를 학습하며, 대량의 데이터와 복잡한 비선형 관계를 처리하는 데 강점이 있다. 
   머신러닝은 상대적으로 간단한 알고리즘으로 이루어져 있으며, 특징(feature)을 사람이 직접 설계하는 경우가 많다. 
   딥러닝은 데이터를 계층적으로 분석하고 특징을 자동으로 추출할 수 있다는 점에서 차별화된다.

## 1. AI 모델링 프로세스  

### 1-(1) 데이터 확인

- 데이터의 크기
- 결측치
- 이상치

### 1-(2) 데이터 전처리 
- 결측치 처리 : 평균, 중앙값 채우거나 삭제 
- 이상치 처리 : 이상값을 수정하거나 삭제 
- 데이터 변환 : 정규화, 표준화 
- 범주형 데이터 인코딩 : 텍스트 형식 -> 숫자
중요성 : 데이터 전처리를 통해 노이즈를 줄이고, 데이터의 품질을 높여 모델의 성능을 극대화할 수 있습니다.

### 1-(3) AI 모델 선정
(3)-1 문제 유형에 따라 모델을 선택 
  - 분류 문제 : 이미지 분류, 이메일 스팸 분류 등
  - 추천 모델 : 결정 트리, 랜덤 포레스트, 로지스틱 회귀.
  - 회귀 문제 : 주택 가격 예측, 매출 예측 등.
  - 추천 모델 : 선형 회귀, 랜덤 포레스트 회귀.
  - 군집화 문제 : 고객 세그먼트 분류.
  - 추천 모델 : K-Means, 계층적 군집화.
(3)-2 데이터 크기와 특성에 따라 모델을 선택합니다.
(3)-3 중용성 : 적합한 모델을 선택하지 않으면 문제 해결에 필요한 성능을 얻기 어려울 수 있습니다. 문제와 데이터에 적합한 모델을 선택하는 것이 핵심입니다.
