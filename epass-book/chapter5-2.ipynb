### (3) 인덱스와 컬럼 계층 변경 

```python
import pandas as pd 

file_path = 'datasets/heart.csv'
df = pd.read_csv(file_path) 

df.columns = ['나이', '성별', '흉통유형', '안정시혈압', '콜레스테롤', '공복혈당', 
              '심전도', '최대심박수', '운동유발협심증', 'ST우울증', '운동후ST검사', 
              '주요혈관수', '결합유형', '심장질환여부']

df.head()
```

```python
grouped = df.groupby(['성별', '흉통유형']).mean()
grouped
```

```python
grouped_reset = grouped.reset_index()
grouped_reset
```

```python
pivot = df.pivot_table(index='성별', columns=['흉통유형', '운동후ST검사'], values='최대심박수', aggfunc='mean')
pivot
```

```python
pivot_reset = pivot.reset_index()
pivot_reset.columns = ['_'.join(map(str, col)).strip() for col in pivot_reset.columns]
pivot_reset
```
